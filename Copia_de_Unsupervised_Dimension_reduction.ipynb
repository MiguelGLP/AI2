#**Dimension reduction** in unsupervised learning refers to the process of reducing the number of random variables under consideration, by obtaining a set of principal variables. It is often used to simplify models, reduce computation time, and remove noise from data while retaining the essential structure and relationships.**

Key techniques for dimension reduction in unsupervised learning include:

# **Principal Component Analysis (PCA)**: This method transforms the original variables into a new set of uncorrelated variables called principal components, which are ordered by the amount of variance they explain in the data. The first few components capture most of the variation, allowing for a reduction in dimensions.

# In the provided example using the Iris dataset, the original dataset has 4 features (or parameters):


*   Sepal length
*   Sepal width
*   Petal length
*   Petal width



# These 4 features will be reduced to 2 principal components using Principal Component Analysis (PCA). The PCA class from scikit-learn is used to perform this dimensionality reduction. Hereâ€™s a more detailed breakdown of the steps and the relevant parameters:




# PCA example with Iris Data-set

Principal Component Analysis applied to the Iris dataset.

See [here](https://en.wikipedia.org/wiki/Iris_flower_data_set) for more
information on this dataset.


```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris # loading data which is embeded in Sklearn as example
import sklearn.decomposition as sd # importing the functions of dimension reduction ( sub_library)

#Load the Iris dataset
data = load_iris()
X = data.data # X can be a detaframe with all parameters which you want to reduce by PCA and these parameters are columns
y = data.target
```

```python
# Apply PCA to reduce the dataset to 2 dimensions ( variables)
pca = sd.PCA(n_components=2) # creat model which means you decide the function ( PCA) and the number of components

X_reduced = pca.fit_transform(X) #Result: X_reduced is the transformed dataset with only 2 features (principal components).
```

```python
# Plot the reduced dataset
plt.figure(figsize=(11, 7))
plt.scatter(X_reduced[:, 1], X_reduced[:, 0], c=y)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
#plt.legend(labels=data.target_names)
plt.show()
```

The fit_transform method of the PCA class computes the principal components and transforms the original 4-dimensional data into a new 2-dimensional space. These two new dimensions are linear combinations of the original four dimensions, chosen to maximize the variance captured from the original data.

The resulting X_reduced contains the coordinates of the original data points in the new 2-dimensional space defined by the first two principal components. Each of these principal components is a weighted combination of the original four features. This means that while we have reduced the dimensionality, the new components still retain as much variance as possible from the original features.

# **Principal Component Analysis (PCA) is considered a method used in machine learning, specifically under the category of unsupervised learning. However, PCA is not an iterative method in the traditional sense used by many optimization algorithms. Hereâ€™s a detailed explanation:**

Machine Learning Aspect
Unsupervised Learning: PCA is an unsupervised learning technique because it does not require labeled data. It finds patterns and structures in the data without reference to known outcomes or labels.
Feature Extraction: PCA is commonly used for feature extraction and dimensionality reduction, helping to simplify models, reduce computation, and mitigate the curse of dimensionality.
Iterative Nature
Non-Iterative Algorithm: Traditional PCA is not inherently iterative like some optimization algorithms (e.g., gradient descent). The standard PCA algorithm typically involves:
Centering the Data: Subtracting the mean of each feature from the dataset.
Computing the Covariance Matrix: Finding the covariance matrix of the centered data.
Eigen Decomposition: Performing an eigen decomposition of the covariance matrix to find the eigenvalues and eigenvectors.
Forming Principal Components: Ordering the eigenvectors by their corresponding eigenvalues in descending order and selecting the top components based on the desired dimensionality.


# **Explained Variance Ratio: The explained variance ratio of the selected principal components is printed. This indicates the proportion of the dataset's variance that is captured by each principal component.**

# **Explained** Variance Ratio
# The explained variance ratio tells you how much information (variance) can be attributed to each of the principal components. A higher explained variance ratio means that the component captures more of the total variability in the data. When PCA is used for dimensionality reduction, you typically want to retain enough components to explain a large portion of the variance while reducing the dimensionality.

In summary, while PCA itself does not have an accuracy metric like supervised learning models, the explained variance ratio is a key metric to understand how well PCA is capturing the variability in the data.

# **Why** Standardization is Important
# Standardizing the data ensures that each feature contributes equally to the analysis. In many machine learning algorithms, features with larger ranges can dominate the objective function, leading to skewed results. Standardization ensures that each feature has a mean of 0 and a variance of 1, allowing algorithms to converge faster and often leading to better performance.

# This step standardizes the features such that each feature has a mean of 0 and a standard deviation of 1. After this transformation, the values are typically within the range of -3 sigma to 3 sigma, but they can occasionally be slightly outside this range, especially if the original data has outliers.


```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Standardize the features
scaler = StandardScaler() # define model
X_scaled = scaler.fit_transform(X) # this is the function to show the explanation-power of the new set of the components in comparison to the original number of the variables

pca_full = PCA().fit(X_scaled)
plt.figure(figsize=(8, 6))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Principal Components')
plt.grid()
plt.show()
```

# **Iterative** Extensions
# Iterative PCA Variants: There are iterative methods and extensions of PCA, such as Incremental PCA, which can handle large datasets by processing data in mini-batches and updating the principal components iteratively. These methods are particularly useful when dealing with streaming data or datasets that do not fit into memory.
## **Incremental PCA (Iterative Variant)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import IncrementalPCA

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Apply Incremental PCA to reduce the dataset to 2 dimensions
n_batches = 10
inc_pca = IncrementalPCA(n_components=2)
for X_batch in np.array_split(X, n_batches):
    inc_pca.partial_fit(X_batch)

X_reduced = inc_pca.transform(X)

# Plot the reduced dataset
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Incremental PCA of Iris Dataset')
#plt.legend(handles=scatter.legend_elements()[0], labels=data.target_names)
plt.show()
```
